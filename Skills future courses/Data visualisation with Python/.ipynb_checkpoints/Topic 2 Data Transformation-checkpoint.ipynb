```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import sklearn

%config Completer.use_jedi = False
```

```python
# Concatenating Data Demo

mtcars_sample = pd.read_csv("mtcars.csv", index_col="car_names")


merc = [c for c in mtcars_sample.index if 'Merc' in c]
merc
```

```python
merc_cars = mtcars_sample.loc[merc]
merc_cars
```

```python
toyota = [c for c in mtcars_sample.index if 'Toyota' in c]
toyota_cars = mtcars_sample.loc[toyota]
toyota_cars
```

```python
merc_toyota_cars = pd.concat([merc_cars, toyota_cars])
merc_toyota_cars
```

```python
# Activity: Selecting and Slicing Data

air_quality_no2 = pd.read_csv("no2.csv")
air_quality_no2 = air_quality_no2[["date.utc", "location","parameter", "value"]]
air_quality_no2
```

```python
air_quality_pm25 = pd.read_csv("pm2.5.csv")
air_quality_pm25 = air_quality_pm25[["date.utc", "location","parameter", "value"]]
air_quality_pm25
```

```python
combined_air = pd.concat([air_quality_no2, air_quality_pm25])
combined_air
```

```python
# Appending Data

toyota_merc_cars2= toyota_cars.append(merc_cars)
toyota_merc_cars2
```

```python
# Merging Data Demo

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K4'],
                      'C': ['C0', 'C1', 'C2', 'C4'],
                      'D': ['D0', 'D1', 'D2', 'D4']})

result = pd.merge(left, right, on='key',how='inner')
result
```

```python
# Activity: Merging Data

new_air = pd.merge(air_quality_no2,air_quality_pm25, on = ['date.utc','location'])
new_air
```

```python
new_air.to_csv("new_air.csv")
```
