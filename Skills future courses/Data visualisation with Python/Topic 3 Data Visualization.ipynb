```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import sklearn

%config Completer.use_jedi = False
```

```python
mtcars_sample = pd.read_csv("mtcars.csv", index_col="car_names")
```

```python
# Create Plots in Pandas

mtcars_sample.plot(figsize=(15,10), title = "MTCars Dataset", grid=True, kind='hist')
```

```python
# Activity: Line Plot

health = pd.read_csv("health_exp.csv",usecols=[1,2,3])
health.plot()
```

```python
# Scatter Plot
# Alpha is opacity

mtcars_sample.plot(x="hp", y="mpg",kind='scatter',alpha=1)

mtcars_sample.plot.scatter(x="hp", y="mpg",alpha=0.5, color='red')
```

```python
# Bar Plot

mtcars_sample.pivot(columns='cyl',values='hp').mean().plot.bar()
```

```python
mtcars_sample.pivot(columns='cyl',values='hp').mean().plot.bar(color='red')
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
ax = df.plot.bar(rot=0)
```

```python
# Horizontal Bar Plot

mtcars_cyl= mtcars_sample.pivot(columns='cyl',values='hp').mean().plot.barh()
```

```python
# Stacked Bar Plot

mtcars_sample.pivot_table(index='am',columns='cyl',values='mpg',aggfunc=['mean']).plot.bar()
```

```python
mtcars_sample.groupby('am')['mpg','hp'].mean().plot.bar(stacked=True)
```

```python
# Pie Plot

mtcars_cyl= mtcars_sample.pivot(columns='cyl',values='hp').mean().plot.pie()
```

```python
# Box Plot

mtcars_cyl= mtcars_sample.pivot(columns='cyl',values='hp').plot.box()
```

```python
mtcars_cyl= mtcars_sample.pivot(columns='cyl',values='mpg').plot.box()
```

```python
# Area Plot

mtcars_sample2 = mtcars_sample[['cyl','mpg','hp']]
mtcars_sample2.groupby(['cyl']).mean().plot.area()
```

```python
# Histogram

mtcars_sample.mpg.plot.hist(bins=10)
```

```python
# Activity: Data Visualization

care_facilities = pd.read_csv("care_facility.csv")
care = care_facilities.pivot(columns='sector',values='count').sum().plot.barh()
```

```python
# Subplot

mtcars_cyl= mtcars_sample.groupby('cyl').mean()
mtcars_cyl.plot(subplots=True,figsize=(13,10))
```

```python
# Activity: Data Visualization

health = pd.read_csv('health_exp.csv')

health_sub = health.groupby('financial_year').mean()
health_sub.plot(subplots=True,figsize=(10,7))
```

```python
# Matplotlib

plt.scatter(x="hp", y="mpg", data = mtcars_sample,color='red')
plt.xlabel('hp')
plt.ylabel('mpg')
plt.title('hp vs mpg')
```

```python
# Seaborn

mtcars_cyl= mtcars_sample.groupby('cyl')['mpg','hp','am'].mean()
sb.heatmap(mtcars_cyl, annot=True)
```
